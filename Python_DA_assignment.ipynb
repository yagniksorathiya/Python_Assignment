7.Write a Python program to get the Fibonacci series between 0 to 50.

Note : The Fibonacci Sequence is the series of numbers :0, 1, 1, 2, 3, 5, 8, 13, 21, ....

Every next number is found by adding up the two numbers before it.

```python
x,y = 0,1
print("Fibonacci sequence:")
while y<50:
    print(y)
    x,y = y, x+y
```

5.Write a python program to scrape reviews from a commercial web site

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_soup(u):
    r = requests.get(u,headers=HEADERS)
    soup = BeautifulSoup(r.text, 'html.parser')
    return soup
review_list=[]
```

```python
def get_reviews(soup):
    reviews = soup.find_all('div', {'data-hook': 'review'})
    try:
        for item in reviews:
            review = {
            'title': item.find('a', {'data-hook': 'review-title'}).text.strip(),
            'rating':float(item.find('i', {"data-hook":"review-star-rating"}).text.replace('out of 5 stars', '').strip()),
            'body': item.find('span', {'data-hook':"review-body"}).text.strip(),
            }
            review_list.append(review)
    except:
        pass
HEADERS = ({'User-Agent':
            'Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
            AppleWebKit/537.36 (KHTML, like Gecko) \
            Chrome/90.0.4430.212 Safari/537.36',
            'Accept-Language': 'en-US, en;q=0.5'})
for x in range(0, 10):
    soup = get_soup(f'https://www.amazon.com/SAMSUNG-Factory-Unlocked-Android-Smartphone/dp/B0BLP3ZZXT/ref=sr_1_2?keywords=iphone%2B14%2Bphone%2Bnew&qid=1695497478&s=wireless&sr=1-2&th=1#customerReviews{x}')
    # print(soup)
    print(f'Getting page: {x}')
    get_reviews(soup)
    print(len(review_list))
    if not soup.find('li', {'class': 'a-disabled a-last'}):
        pass
    else:
        break
```

```python
final_list = pd.DataFrame(review_list)
final_list
```
